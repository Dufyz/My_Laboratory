```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

import numpy as pd

import pandas as pd
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
```

```python
plt.style.use('default')
```

```python
data = load_iris()
```

```python
x_train, x_test, y_train, y_test = train_test_split(data['data'], data['target'], random_state = 0)
```

```python
print(f'Treino {x_train.shape}') # 75% treino
print(f'Treino {x_test.shape}') # 25% teste
```

```python
df = pd.DataFrame(x_train, columns = data.feature_names)
grr = scatter_matrix(df, c = y_train, figsize = (15,15), marker = 'o', 
                     hist_kwds = {'bins': 30}, s = 60, alpha = 8)
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(x_train, y_train)
```

```python
print(f'Treino : {knn.score(x_train, y_train)}') 
print(f'Teste : {knn.score(x_test, y_test)}')  
```
