```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
age_x = [1, 2, 3, 4, 5]
x_index = np.arange(len(age_x))
width = 0.10
```

```python
dev_y = [1, 3, 6, 8, 10]
plt.bar(x_index + (width), dev_y, width = width, color = 'orange', label = 'Dev')

doctor_y = [3, 5, 6, 7, 13]
plt.bar(x_index + (2* width), doctor_y, width = width, color = 'red', label = 'Doctor' )

engineer_y = [4, 4, 6, 8, 11]
plt.bar(x_index + (3 * width), engineer_y, width = width, color = 'pink', label = 'Engineer')

journalist_y = [2, 1, 4, 3 ,4]
plt.bar(x_index + (4 * width) , journalist_y, width = width, color = 'blue', label = 'Journalist')

plt.xlabel('Age')
plt.ylabel('Salary')

plt.legend(loc =  2)
plt.grid(True)
#plt.xticks(ticks = x_index, labels = age_x)
#plt.savefig('plot2.png')
```

```python
plt.plot(age_x, dev_y, marker = 'o', markersize = 10, color = 'orange', label = 'Dev')
plt.plot(age_x, doctor_y, marker = 'o', markersize = 10, color = 'red', label = 'Doctor')
plt.plot(age_x, engineer_y, marker = 'o', markersize = 10, color = 'pink', label = 'Engineer')
plt.plot(age_x, journalist_y, marker = 'o', markersize = 10, color = 'blue', label= 'Journalist')

plt.xlabel('Age')
plt.ylabel('Salary')

plt.legend(loc = 0)
plt.grid(True)
#plt.savefig('plot_salary_by_profission.png')
```
