# Figure Objects

```python
plt.style.use('fivethirtyeight')
#plt.style.use('ggplot')

fig_1 = plt.figure(figsize = (5,4), dpi = 100)
axes_1 = fig_1.add_axes([0.1, 0.2, 0.9, 0.9])

#axes_1.set_xlabel('X')
#axes_1.set_ylabel('Y')

axes_1.plot(x_1, y_1, '--', marker = '.', linewidth = 1, label = 'X', color = 'purple')
axes_1.plot(y_1, x_1, marker = 'o', linewidth = 1,label = 'Y', color = 'orange')

axes_1.grid(True)
axes_1.legend(loc = 0)

#plt.savefig('plot.png')
```
